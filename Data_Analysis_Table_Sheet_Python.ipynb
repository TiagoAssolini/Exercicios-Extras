```python
prof = pd.read_excel('/content/Dados_CursoEstatistica.xlsx', sheet_name = 3)
prof.head()
```

```python
# Mudanca de nome de COLUNA pra qualquer texto
prof.columns = ['professir','vergonha','Exp']
```

```python
prof.describe()
```

```python
prof.boxplot()
```

```python
prof.boxplot('vergonha')
```

```python
prof.hist('Exp')
```

```python
prof.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(prof['vergonha'], prof['Exp'])
plt.show()
```

```python
#invertendo
plt.scatter(prof['Exp'], prof['vergonha'])
plt.title('Exp x Vergonha')
plt.xlabel('Exp')
plt.ylabel('Vergonha')
plt.show()
```

```python
sns.scatterplot(x="Exp", y="vergonha", data=prof)
plt.title('Exp x Vergonha')
plt.xlabel('Exp')
plt.ylabel('Vergonha')
plt.show()
```
